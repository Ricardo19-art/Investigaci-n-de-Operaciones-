# <font color="RoyalBlue"> Introduccion a Networkx</font>

Networkx es una libreria de python la cual es idonea para trabajar con grafos(redes) aun nivel medio-avanzado con un manejor relativamnte facil de sus comandos.

**NetworX nos ofrece cuatro clases de grafos**

*   Graph: Grafo no dirigido
*   Digraph: Grafo dirigido
*   Multigraph: Grafo no dirigido con múltiples aristas entre nodos
*   MultiDigraph: Grafo dirigido con múltiples aristas



### <font color="SeaGreen"> Y que es un grafo?</font>

Este se donomina como una estructura de datos interconectados que consiste de **nodos(puntos)** conectectados mediante **aristas(lineas)**.

### <font color="Purple"> Importamos nuestras librerias </font>

```python
import networkx as nx  # Importa la libreria de networkx y le asigna una abreviatura (nx)
import matplotlib.pyplot as plt   #Importa la libreria matplotlib para graficar de mejor manera en conjunto con networkx
```

# <font color="Crimson"> Creamos nuestro grafo no dirigido</font>



```python
G_nd = nx.Graph()     # Crea un grafo vacio
```

### <font color="DarkOrange"> Creamos nuestros nodos (O, P y Q) y los unimos formando nuestra red</font>

```python
G_nd.add_edge("O", "P")   # Genera un arco que une O y P
G_nd.add_edge("P", "Q")   # Genera un arco que une P y R
G_nd.add_edge("Q", "O")   # Genera un arco que une R y O
```

### <font color="DarkOrange"> Obtenemos el tipo de gráfico que es G</font>

```python
G_nd   # Nos regresa el tipo de gráfico
```

### <font color="DarkOrange"> Imprimimos nuestra gráfica con nuestros nodos y sus respectivos arcos</font>

```python
nx.draw_networkx(G_nd, with_labels=True,   # with_labes hace que se vean los nombres de los nodos
                 node_color="lightblue")   # Le da el color deseado a nuestros nodos
plt.title("Grafo no dirigido")       # Sirve para colocarle el nombre a la gráfica
plt.show()                                 # Muestra el grafo
```

### <font color="DarkOrange"> Obtenemos todos los nodos de nuestra red</font>

```python
nx.nodes(G_nd)   # Nos regresa los nodos de la red

print("Nodos:", nx.nodes(G_nd))   # Imprimimos los ndoos de la red
```

### <font color="DarkOrange"> Obtenemos todos los arcos con sus respectivos nodos de nuestra red</font>

```python
nx.edges(G_nd)   # Nos regresa los arcos entre cada nodo de la red

print("Arcos:", nx.edges(G_nd))   # Imprimios los arcos entre cada nod de la red
```

### <font color="DarkOrange"> Obtenemos el número total de nodos</font>

```python
nx.number_of_nodes(G_nd)   # Nos regresa el número de nodos

print("Número de nodos:", nx.number_of_nodes(G_nd))    # Imprime el número total de nodos
```

### <font color="DarkOrange"> Obtenemos el número total de arcos</font>

```python
nx.number_of_edges(G_nd)    # Nos regresa el número de arcos

print("Número de arcos:", nx.number_of_edges(G_nd))    # Imprime el número total de arcos
```

### <font color="DarkOrange"> Obtenemos el número de conexiones por nodo </font>


```python
dict(G_nd.degree)

print("Numero de conexiones por nodo:", dict(G_nd.degree))
```

# <font color="Crimson"> Arbol de expanción minima</font>





```python
G1 = nx.Graph()                        # Creamos un grafo vacio

G1.add_weighted_edges_from([
    (1, 2, 4), (1, 3, 1), (2, 3, 2),     # Añadimos los vertices, primeros 2 valores y sus respectivos arcos con sus longitudes, ultimo valor (vertice, vertice, arista)
    (2, 4, 5), (3, 4, 8), (3, 5, 3)
])

arbol_minimo = nx.minimum_spanning_tree(G1)      # Obtenemos nuestro árbol de expansión mínima con (nx.minimum_spanning_tree) de nuestro grafo(G1)

print("Aristas del árbol de expansión mínima:")
for nodo1, nodo2, valor in arbol_minimo.edges(data="weight"):     # Imprimimos las aristas del nuestro árbol con sus respectivos valores
    print(nodo1, "-", nodo2, "valor:", valor)

posiciones = nx.spring_layout(G1, seed=42)         # Usamos un diseño fijo para que ambos grafos tengan la misma distribución

# Grafo original
plt.figure()
nx.draw(G1, posiciones, with_labels=True, node_color="lightyellow")
nx.draw_networkx_edge_labels(G1, posiciones, edge_labels=nx.get_edge_attributes(G1, "weight"))
plt.title("Grafo original")
plt.show()

# Árbol de expansión mínima
plt.figure()
nx.draw(arbol_minimo, posiciones, with_labels=True, node_color="lightgreen")
nx.draw_networkx_edge_labels(arbol_minimo, posiciones, edge_labels=nx.get_edge_attributes(arbol_minimo, "weight"))
plt.title("Árbol de expansión mínima")
plt.show()
```

# <font color="Crimson"> Ruta más corta</font>

```python
G2 = nx.Graph()                                        # Creamos un Grafo vacio

G2.add_weighted_edges_from([
    ('A', 'B', 4), ('A', 'C', 2), ('B', 'C', 1),
    ('B', 'D', 5), ('C', 'D', 8), ('C', 'E', 10),      # Añadimos los vertices, primeros 2 valores y sus respectivos arcos con sus longitudes, ultimo valor (vertice, vertice, arista)
    ('D', 'E', 2)
])

ruta = nx.shortest_path(G2, source='A', target='E', weight='weight')               # La función nx.shortest nos calcula la ruta mas corta solo debemos dar el punto "A" y "B" osea de donde a donde
distancia = nx.shortest_path_length(G2, source='A', target='E', weight='weight')   # La funcion nx.shortest_path_length nos calcula la distancia de la ruta mas corta

print(f"Ruta más corta de A a E: {' -> '.join(ruta)}")       # Imprime el texto de la "Ruta mas corta de A a E" y nos arroja la sucesion de nodos de la ruta mas corta
print(f"Distancia total: {distancia}")                       # Imprime la distancia de la ruta mas corta

posiciones = nx.spring_layout(G2, seed=42)                # Posiciones para dibujar el grafo

# Grafo completo con etiquetas de valor
plt.figure()
nx.draw(G2, posiciones, with_labels=True, node_color='lightyellow')
etiquetas = nx.get_edge_attributes(G2, 'weight')
nx.draw_networkx_edge_labels(G2, posiciones, edge_labels=etiquetas)

# Resaltar en rojo la ruta más corta
aristas_ruta = list(zip(ruta[:-1], ruta[1:]))
nx.draw_networkx_edges(G2, posiciones, edgelist=aristas_ruta, edge_color='red', width=3)
nx.draw_networkx_nodes(G2, posiciones, nodelist=ruta, node_color='red')

plt.title(f"Ruta más corta: A → E (Distancia: {distancia})")
plt.show()
```

# <font color="Crimson"> Flujo máximo</font>

```python
# Hacemos el grafo dirigido con aristas que tienen capacidad de flujo
G3 = nx.DiGraph()                          # Creamos un grafo vacio

# Para crear nuestro grafo dirigido necesitamos proporcionr arcos dirigidos entre nuestros nodos
G3.add_edge('Origen', 'A', capacity=3)     # Nodo Origen a Nodo A con capacidad de 3
G3.add_edge('Origen', 'B', capacity=2)     # Nodo Origen a Nodo B con capacidad de 2
G3.add_edge('A', 'C', capacity=3)          # Nodo A a Nodo C con capacidad de 3
G3.add_edge('B', 'C', capacity=2)          # Nodo B a Nodo C con capacidad de 2
G3.add_edge('B', 'D', capacity=2)          # Nodo B a Nodo D con capacidad de 2
G3.add_edge('C', 'Destino', capacity=4)    # Nodo C a Nodo Destino con capacidad 4
G3.add_edge('D', 'Destino', capacity=3)    # Nodo D a Nodo Destino con capacidad 3

valor_maximo, flujo = nx.maximum_flow(G3, 'Origen', 'Destino')    # Calcular el flujo máximo entre origen y destino

# Mostrar el valor total y el flujo en cada arista con capacidad
print(f"Valor del flujo máximo: {valor_maximo}")
print("Flujo en cada arista:")
for nodo_inicio, vecinos in flujo.items():
    for nodo_fin, cantidad in vecinos.items():
        if cantidad > 0:
            capacidad = G3[nodo_inicio][nodo_fin]['capacity']
            print(f"  {nodo_inicio} → {nodo_fin}: {cantidad}/{capacidad}")

posiciones = nx.spring_layout(G3, seed=42)   # Le da sus respectivas posiciones a nodo

etiquetas = {(u, v): f"{flujo[u].get(v, 0)}/{G3[u][v]['capacity']}" for u, v in G3.edges()}   # Etiquetas para cada arista con relación flujo/capacidad

# Dibujar el grafo con las capacidades y el flujo calculado
plt.figure()
nx.draw(G3, posiciones, with_labels=True, node_color='lightcoral', arrows=True)
nx.draw_networkx_edge_labels(G3, posiciones, edge_labels=etiquetas)
plt.title(f"Flujo Máximo: {valor_maximo}")
plt.show()
```
